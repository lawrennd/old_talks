```python
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(-4, 4, 50)
```

```python
y = 1./(1+np.exp(-x))
```

```python
y2 = 1./(1+np.exp(-(x-2)))
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, y, 'y', linewidth=3)
ax.plot(x, y2, 'c', linewidth=3)
plt.axis('off')
plt.savefig('slides/diagrams/data-science-ai-time-frame.svg')
```
